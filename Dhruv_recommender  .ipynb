```
with open("Dhruv_Bread_topic_names_1hack.pickle", "rb") as input_file:
   temp_topic_names = pickle.load(input_file)

topic_names = [x[1] for x in temp_topic_names]
```

```
with open("Dhruv_Bread_text_submissions_1hack.pickle", "rb") as input_file:
   temp_submissions = pickle.load(input_file)

submissions = [x[1] for x in temp_submissions]
```

```
embedings = embedder.encode(submissions)
```

```
#value of 1.0 means exactly the same
distances = []
for vector1 in embedings:
  temp = []
  for vector2 in embedings:
    dist = scipy.spatial.distance.cdist([vector1], [vector2], "cosine")
    temp.append(1.0 - distance)
  distances.append(temp)
```

```
def Nmaxelements(list1, N): 
  final_list = [] 
  list1_copy = list1.copy() 

  for i in range(0, N):  
    max1 = 0
    for j in range(len(list1_copy)):      
      if list1_copy[j] > max1: 
        max1 = list1_copy[j]; 
              
    list1_copy.remove(max1); 
    final_list.append([i,max1,topic_names[i],submissions[i]]) 

  return(final_list[1:])
```

```
recommendations = []
for i in range(len(submissions)):
  recommendations.append(Nmaxelements(distances[i],6))
  

```
